```python
# For Gender
import numpy as np


def std(x):
    return np.std(x)



def build_tree(dff, op):

    dff

    split_col = None
    max_std = -1

    y_std = np.std(dff[op])

    if std(dff[op]) * std(dff[op]) > 0.1:
        
        return dff[op].mean()
    
    att_crit = []
    
    for col in dff.keys():



        if col != op:
 
          

            tmp_df = dff.filter(items=[col, op])
            
            x = tmp_df.groupby(col).agg([std, "count"])


            x.columns = ["std", "count"]

            print(x)


            sd_x_y = 0
            for u in tmp_df[col].unique():

                sd_x_y += x["std"][u] * (x["count"][u] / len(tmp_df.index))
            
            if y_std - sd_x_y > max_std:
                max_std = y_std - sd_x_y
                att_crit = tmp_df[col].unique()
                split_col = col

    
    for att in att_crit:
        dff = dff[dff[split_col] == att].filter(items=[x for x in dff.keys() if x != split_col])
        return build_tree(dff, op)


    
    
    



        

build_tree(df, "HoursPlayed")
```
