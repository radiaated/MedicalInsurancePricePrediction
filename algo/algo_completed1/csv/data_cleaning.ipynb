```python
import pandas as pd

df = pd.read_csv("medidataset2.csv")

df.describe()

```

```python
for i in df.index:
    if 18 <= df.loc[i, "age"] < 26:
        df.loc[i, "age2"] = "18-26"
    elif 26 <= df.loc[i, "age"] < 41:
        df.loc[i, "age2"] = "26-41"
    elif 41 <= df.loc[i, "age"] < 56:
        df.loc[i, "age2"] = "41-56"
    elif df.loc[i, "age"] >= 56:
        df.loc[i, "age2"] = ">56"

df.drop(columns="age", inplace=True)

df.rename(columns={'age2': 'age'}, inplace=True)
```

```python
for i in df.index:
    if df.loc[i, "bmi"] < 18.5:
        df.loc[i, "bmi2"] = "< 18.5"
    elif 18.5 <= df.loc[i, "bmi"] < 25:
        df.loc[i, "bmi2"] = "18.5 - 25"
    elif 25 <= df.loc[i, "bmi"] < 30:
        df.loc[i, "bmi2"] = "25 - 30"
    elif 30 <= df.loc[i, "bmi"] < 35:
        df.loc[i, "bmi2"] = "30 - 35"
    elif 35 <= df.loc[i, "bmi"] < 40:
        df.loc[i, "bmi2"] = "35 - 40"
    elif df.loc[i, "bmi"] >= 40:
        df.loc[i, "bmi2"] = ">=40"
    

df.drop(columns="bmi", inplace=True)

df.rename(columns={'bmi2': 'bmi'}, inplace=True)
```

```python
for i in df.index:
    if df.loc[i, "sex"] == "male":
        df.loc[i, "sex2"] = 0
    elif df.loc[i, "sex"] == "female":
        df.loc[i, "sex2"] = 1

df.drop(columns="sex", inplace=True)

df.rename(columns={'sex2': 'sex'}, inplace=True)

df.describe()
```

```python
for i in df.index:
    if df.loc[i, "age"] == "18-26":
        df.loc[i, "age2"] = 0
    elif df.loc[i, "age"] == "26-41":
        df.loc[i, "age2"] = 1
    elif df.loc[i, "age"] == "41-56":
        df.loc[i, "age2"] = 2
    elif df.loc[i, "age"] == ">56":
        df.loc[i, "age2"] = 3

df.drop(columns="age", inplace=True)

df.rename(columns={'age2': 'age'}, inplace=True)

df.describe()
```

```python

for i in df.index:
    if df.loc[i, "bmi"] == "< 18.5":
        df.loc[i, "bmi2"] = 0
    elif df.loc[i, "bmi"] == "18.5 - 25":
        df.loc[i, "bmi2"] = 1
    elif df.loc[i, "bmi"] == "25 - 30":
        df.loc[i, "bmi2"] = 2
    elif df.loc[i, "bmi"] == "30 - 35":
        df.loc[i, "bmi2"] = 3
    elif df.loc[i, "bmi"] == "35 - 40":
        df.loc[i, "bmi2"] = 4
    elif df.loc[i, "bmi"] == ">=40":
        df.loc[i, "bmi2"] = 5
    

df.drop(columns="bmi", inplace=True)

df.rename(columns={'bmi2': 'bmi'}, inplace=True)

df
```

```python
for i in df.index:
    if df.loc[i, "smoker"] == "yes":
        df.loc[i, "smoker2"] = 0
    elif df.loc[i, "smoker"] == "no":
        df.loc[i, "smoker2"] = 1

df.drop(columns="smoker", inplace=True)

df.rename(columns={'smoker2': 'smoker'}, inplace=True)

df.describe()
```

```python
for i in df.index:
    if df.loc[i, "region"] == 'southwest':
        df.loc[i, "region2"] = 0
    elif df.loc[i, "region"] == 'southeast':
        df.loc[i, "region2"] = 1
    elif df.loc[i, "region"] == 'northwest':
        df.loc[i, "region2"] = 2
    elif df.loc[i, "region"] == 'northeast':
        df.loc[i, "region2"] = 3

df.drop(columns="region", inplace=True)

df.rename(columns={'region2': 'region'}, inplace=True)

df.describe()
```

```python
for i in df.index:
    if df.loc[i, "occupation"] == 'Student':
        df.loc[i, "occupation2"] = 0
    elif df.loc[i, "occupation"] == 'White collar':
        df.loc[i, "occupation2"] = 1
    elif df.loc[i, "occupation"] == 'Unemployed':
        df.loc[i, "occupation2"] = 2
    elif df.loc[i, "occupation"] == 'Blue collar':
        df.loc[i, "occupation2"] = 3

df.drop(columns="occupation", inplace=True)

df.rename(columns={'occupation2': 'occupation'}, inplace=True)

df.describe()

```
