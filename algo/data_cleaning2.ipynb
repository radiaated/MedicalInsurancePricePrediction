```python
import pandas as pd

df = pd.read_csv("./df_reduced.csv")

df.dropna(inplace=True)

for i in df.index:
    if 18 <= df.loc[i, "age"] < 26:
        df.loc[i, "age2"] = "18-26"
    elif 26 <= df.loc[i, "age"] < 41:
        df.loc[i, "age2"] = "26-41"
    elif 41 <= df.loc[i, "age"] < 56:
        df.loc[i, "age2"] = "41-56"
    elif df.loc[i, "age"] >= 56:
        df.loc[i, "age2"] = ">56"

df.drop(columns="age", inplace=True)

df.rename(columns={'age2': 'age'}, inplace=True)

df
```

```python

for i in df.index:
    if df.loc[i, "bmi"] < 18.5:
        df.loc[i, "bmi2"] = "< 18.5"
    elif 18.5 <= df.loc[i, "bmi"] < 25:
        df.loc[i, "bmi2"] = "18.5 - 25"
    elif 25 <= df.loc[i, "bmi"] < 30:
        df.loc[i, "bmi2"] = "25 - 30"
    elif 30 <= df.loc[i, "bmi"] < 35:
        df.loc[i, "bmi2"] = "30 - 35"
    elif 35 <= df.loc[i, "bmi"] < 40:
        df.loc[i, "bmi2"] = "35 - 40"
    elif df.loc[i, "bmi"] >= 40:
        df.loc[i, "bmi2"] = "40 - 54"
    

df.drop(columns="bmi", inplace=True)

df.rename(columns={'bmi2': 'bmi'}, inplace=True)

df
```

```python
df['gender'] = df['gender'].str.strip()
df['smoker'] = df['smoker'].str.strip()
df['medical_history'] = df['medical_history'].str.strip()
df['family_medical_history'] = df['family_medical_history'].str.strip()
df['exercise_frequency'] = df['exercise_frequency'].str.strip()
df['occupation'] = df['occupation'].str.strip()
df['coverage_level'] = df['coverage_level'].str.strip()
```

```python
df.to_csv("./df_cleaned2.csv", sep=",", index=False)
```
