```python
import numpy as np
import pandas as pd
from sklearn.mixture import GaussianMixture


from sklearn.preprocessing import LabelEncoder


df = pd.read_csv("./Health_insurance.csv")
df
```

```python


# Fit and transform the 'Category' column
df['sex'] = df["sex"].map({"male":0,"female":1})
df['smoker'] = df["smoker"].map({"yes":0,"no":1})
df['region'] = df["region"].map({'southwest':0, 'southeast':1, 'northwest':2, 'northeast':3})
```

```python
from sklearn.neighbors import KernelDensity

# Assuming df is your DataFrame
additional_rows = 5564  # Adjust as needed

# Fit KDE on existing data
kde = KernelDensity(bandwidth=0.2, kernel='gaussian')
kde.fit(df)

# Generate new samples
new_samples = kde.sample(additional_rows)
new_data = pd.DataFrame(new_samples, columns=df.columns)

# Concatenate the new data with the existing DataFrame
augmented_df = pd.concat([df, new_data], ignore_index=True)
```

```python
for k in range(len(augmented_df.index)):
    if augmented_df.loc[k, "sex"] < 0.5:
        augmented_df.loc[k, "sex"] = "male"
    else:
        augmented_df.loc[k, "sex"] = "female"
```

```python
for k in range(len(augmented_df.index)):
    if augmented_df.loc[k, "smoker"] < 0.5:
        augmented_df.loc[k, "smoker"] = "yes"
    else:
        augmented_df.loc[k, "smoker"] = "no"

        
for k in range(len(augmented_df.index)):
    if augmented_df.loc[k, "region"] < 0.5:
        augmented_df.loc[k, "region"] = "southwest"
    elif 0.5 <= augmented_df.loc[k, "region"] < 1.5:
        augmented_df.loc[k, "region"] = "southeast"
    elif 1.5 <= augmented_df.loc[k, "region"] < 2.5:
        augmented_df.loc[k, "region"] = "northwest"
    elif augmented_df.loc[k, "region"] >= 2.5:
        augmented_df.loc[k, "region"] = "northeast"


for k in range(len(augmented_df.index)):
    if augmented_df.loc[k, "children"] < 0.5:
        augmented_df.loc[k, "children"] = 0
    elif 0.5 <= augmented_df.loc[k, "children"] < 1.5:
        augmented_df.loc[k, "children"] = 1
    elif 1.5 <= augmented_df.loc[k, "children"] < 2.5:
        augmented_df.loc[k, "children"] = 2
    elif 2.5 <= augmented_df.loc[k, "children"] < 3.5:
        augmented_df.loc[k, "children"] = 3
    elif 3.5 <= augmented_df.loc[k, "children"] < 4.5:
        augmented_df.loc[k, "children"] = 4
    elif augmented_df.loc[k, "children"] >= 4.5:
        augmented_df.loc[k, "children"] = 5
```
