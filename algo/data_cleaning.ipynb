```python
import pandas as pd
import random

df = pd.read_csv("./csv/aug_csv.csv")
```

```python

for i in range(len(df.index)):
    
    p = random.random()    

    if p < 0.2:
        df.loc[i,"occupation"] = "Unemployed"
        continue

    p = random.random()    

    if df.loc[i,"age"] < 24:
        df.loc[i,"occupation"] = "Student" 
    elif 24 <= df.loc[i,"age"] < 65:
        df.loc[i,"occupation"] = "White collar" if p < 0.9 else "Blue collar" 
```

```python


for i in df.index:
    if 18 <= df.loc[i, "age"] < 26:
        df.loc[i, "age2"] = "18-26"
    elif 26 <= df.loc[i, "age"] < 41:
        df.loc[i, "age2"] = "26-41"
    elif 41 <= df.loc[i, "age"] < 56:
        df.loc[i, "age2"] = "41-56"
    elif 56 <= df.loc[i, "age"] <= 64:
        df.loc[i, "age2"] = "56-64"

df.drop(columns="age", inplace=True)

df.rename(columns={'age2': 'age'}, inplace=True)

df.describe()

     
```

```python

for i in df.index:
    if df.loc[i, "bmi"] < 18.5:
        df.loc[i, "bmi2"] = "< 18.5"
    elif 18.5 <= df.loc[i, "bmi"] < 25:
        df.loc[i, "bmi2"] = "18.5 - 25"
    elif 25 <= df.loc[i, "bmi"] < 30:
        df.loc[i, "bmi2"] = "25 - 30"
    elif 30 <= df.loc[i, "bmi"] < 35:
        df.loc[i, "bmi2"] = "30 - 35"
    elif 35 <= df.loc[i, "bmi"] < 40:
        df.loc[i, "bmi2"] = "35 - 40"
    elif 40 <= df.loc[i, "bmi"] <= 54:
        df.loc[i, "bmi2"] = "40 - 54"
    

df.drop(columns="bmi", inplace=True)

df.rename(columns={'bmi2': 'bmi'}, inplace=True)

df
```
