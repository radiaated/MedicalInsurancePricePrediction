```python
import pandas as pd
import secrets



```

```python
def returnRandom(beg,end):
    
    res = secrets.randbelow(end)
    
    
    
    return res % 20 + 5
```

```python
df = pd.read_csv("Health_insurance.csv")

for i in range(len(df.index)):

    if df.loc[i,"smoker"] == "yes":

        p = secrets.randbelow(10)

        print(p)

        if p <= 3:
            df.loc[i,"diseases"] = "None"
        elif 3 <= p <= 7:
            df.loc[i,"diseases"] = "Heart Disease"
        elif 7 <= p <= 9:
            df.loc[i,"diseases"] = "High Blood Pressure"
        else:
            df.loc[i,"diseases"] = "Diabetes"
    
    else:
        df.loc[i,"diseases"] = "None"
```

```python
for i in range(len(df.index)):

    if df.loc[i,"age"] > 45:

        p = secrets.randbelow(10)

        print(p)

        if p <= 2:
            df.loc[i,"diseases"] = "None"
        elif 2 <= p <= 7:
            df.loc[i,"diseases"] = "Heart Disease"
        elif 7 <= p <= 9:
            df.loc[i,"diseases"] = "High Blood Pressure"
        else:
            df.loc[i,"diseases"] = "Diabetes"
    
    else:
        df.loc[i,"diseases"] = df.loc[i,"diseases"]
```

```python
for i in range(len(df.index)):

    if df.loc[i,"diseases"] !=  "None":
        print("increased")

        df.loc[i,"charges"] = df.loc[i,"charges"] + ((returnRandom(5,20) / 100) * df.loc[i,"charges"])
    
    else:
        df.loc[i,"diseases"] = df.loc[i,"diseases"]
```
