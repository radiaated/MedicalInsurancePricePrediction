```python
import pandas as pd
import secrets
import random



```

```python
def returnRandom(beg,end):
    
    res = secrets.randbelow(end)
    
    
    
    return res % 20 + 5
```

```python
df = pd.read_csv("./csv/aug_csv_cleaned3.csv")

for i in range(len(df.index)):

    if df.loc[i,"smoker"] == "yes":

        p = random.random()

        # print(p)

        if p < 0.3:
            df.loc[i,"medical_history"] = "No"
        elif 0.3 <= p < 0.7:
            df.loc[i,"medical_history"] = "Heart Disease"
        elif 0.7 <= p < 0.9:
            df.loc[i,"medical_history"] = "High Blood Pressure"
        else:
            df.loc[i,"medical_history"] = "Diabetes"
    
    else:
        df.loc[i,"medical_history"] = "No"
```

```python
import math
import importlib
from ID3_classify import ID3
from trees import trees_classify

open(f"./trees/trees_classify.py", "w").close()

dff = df.copy()

dff.drop(columns=["charges"], inplace=True)



yes_count = len(dff[dff["package_Bronze"] == True])
no_count = len(dff[dff["package_Bronze"] == False])

print(yes_count)
print(no_count)

te =  math.log(yes_count/no_count)



prob =  math.exp(te) / ( 1+ math.exp(te))


dff["Prob"] = prob
dff["LogOdds"] = te
for t in range(50):

   

    dff["package_Bronze"] = df["package_Bronze"]



    for k in dff.index:
        if dff.loc[k, "package_Bronze"] == True:
            dff.loc[k, "Resi"] = 1 - dff.loc[k, "Prob"]
        elif dff.loc[k, "package_Bronze"] == False:
            dff.loc[k, "Resi"] = 0 - dff.loc[k, "Prob"]

   
    dff = dff.drop(columns=["package_Bronze"])



    id = ID3(dff,"Resi")
    id.build_tree()
    id.create_tree(f"tree{t}")

    importlib.reload(trees_classify)

    for k in dff.index:
        q = getattr(trees_classify, f"tree{t}")(dff.loc[k].to_dict())
        new_te = dff.loc[k, "LogOdds"] + 0.5 * q
 
        prob = math.exp(new_te) / (1 + math.exp(new_te))
        dff.loc[k, "LogOdds"] = new_te
        dff.loc[k, "Prob"] = prob

dff["package_Bronze"] = df["package_Bronze"]
print(dff)
```

```python
for i in range(len(dff.index)):

    if dff.loc[i,"Prob"] < 0.5:
        dff.loc[i,"Predd"] = False
    elif dff.loc[i,"Prob"] >= 0.5:
        dff.loc[i,"Predd"] = True
```
